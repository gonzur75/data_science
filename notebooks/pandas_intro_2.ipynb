```python
df.groupby("acquisitionYear").size()
```

```python
acquisition_year = df.groupby("acquisitionYear").size()
```

```python
type(acquisition_year)
```

```python
import matplotlib.pyplot as plt

# dane s
acquisition_year.plot()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(x=acquisition_year.index, y=acquisition_year.values, c='goldenrod')
ax.set_yscale('log')
ax.set_title('Number of artworks acquired by year')
ax.set_xlabel("Acquisition year")
ax.set_ylabel("Number of artworks")

ticks = [year for year in acquisition_year.index if year % 10 == 0]

ax.set_xticks(ticks)
ax.set_xticklabels(ticks, rotation=45)
ax.grid()
ax.set_axisbelow(True)
fig.tight_layout()
plt.show()
```
